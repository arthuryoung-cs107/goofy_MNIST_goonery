```python
import autograd.numpy as np
from sklearn.datasets import fetch_openml
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
samples_start = 0
samples_end = 10000
X_subset = X[samples_start:samples_end]
y_subset = y[samples_start:samples_end]
y_subset = y_subset.astype(int)
```

```python
# Visualize some samples from the training dataset
fig, ax = plt.subplots(2, 5, figsize=(25, 10))
for i in range(5):
    ax[0, i].imshow(X_subset[y_subset == i][0].reshape((28, 28)), cmap=plt.cm.gray_r, interpolation='nearest')
    ax[1, i].imshow(X_subset[y_subset == i+5][0].reshape((28, 28)), cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

```python
save_num = samples_end - samples_start
for i in range(save_num):
    pd.DataFrame(X_subset[i].reshape((28, 28))).to_csv("./MNist_csv_raw/MNIST"+ str(samples_start + i) + ".csv")

pd.DataFrame(y_subset[:save_num]).to_csv("./MNist_csv_raw/MNIST_y.csv")
```
